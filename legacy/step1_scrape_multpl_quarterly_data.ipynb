```python
def get_data_from_multpl_website(url):
    headers = {
        'User-Agent': (
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 '
            '(KHTML, like Gecko) Chrome/63.0.3239.108 Safari/537.36'
        )
    }
    res = requests.get(url, headers=headers)
    parsed = HTMLParser.fromstring(res.content.decode('utf-8'))
    tr_elems = parsed.cssselect('#datatable tr')
    raw_data = [[td.text.strip() for td in tr_elem.cssselect('td')] for tr_elem in tr_elems[1:]]
    return raw_data
```

```python
for idx, multpl_dataset in enumerate(multpl_data_desc):
    short_name = multpl_dataset[0]
    long_name = multpl_dataset[1]
    dataset_url = multpl_dataset[2]
    value_type = multpl_dataset[3]
        
    raw_data = get_data_from_multpl_website(dataset_url)

    temp_df = pd.DataFrame([[temp_row[0], temp_row[1]] for temp_row in raw_data], columns=['date', short_name])
    temp_df['date'] = pd.to_datetime(temp_df['date'], format='%b %d, %Y')

    if value_type == 'percent':
        temp_df[short_name] = temp_df[short_name].str.replace('%', '')

    if value_type == 'million':
        temp_df[short_name] = temp_df[short_name].str.replace(' million', '')

    if value_type == 'trillion':
        temp_df[short_name] = temp_df[short_name].str.replace(' trillion', '')

    temp_df[short_name] = temp_df[short_name].replace('', np.nan)

    temp_df[short_name] = pd.to_numeric(temp_df[short_name].str.replace(',', '').astype(float))  # handle commas inside strings

    if value_type == 'percent':
        temp_df[short_name] = temp_df[short_name] / 100.0

    temp_df = temp_df.dropna()
    temp_df = temp_df.set_index('date')
    
    temp_series = temp_df[short_name].resample("QE").last()
    
    if idx == 0:
        multpl_df = temp_series.to_frame()
    else:
        multpl_df = pd.concat([multpl_df, temp_series], axis=1)

multpl_df.columns = [item[0] for item in multpl_data_desc]
multpl_df.index = multpl_df.index.strftime('%Y-%m-%d')
```

```python
multpl_df.to_csv("../data/multpl_datasets_snapshot_" + today_str + ".csv")
multpl_df.to_pickle("../data/multpl_datasets_snapshot_" + today_str + ".pickle")
```
