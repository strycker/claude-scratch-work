```python
import pandas as pd
import os
from datetime import datetime
from dotenv import load_dotenv
from fredapi import Fred
```

```python
today_str = datetime.today().strftime("%Y%m%d")
today_str
```

```python
# Can add the FRED API Key to the line below,
# but best practice is to add a .env file to your git repo containing
# the line FRED_API_KEY="your-api-key"
# this way you have access to the key on your local area for execution
# but keep it out of the public github repo

# fred_api_key = "MY_FRED_API_KEY_HERE"

# Load environment variables from .env file
load_dotenv()

# Access the API key
fred_api_key = os.getenv("FRED_API_KEY")

# Use the key in your application
# if fred_api_key is not None:
    # print(f"My FRED API Key is: {fred_api_key}")
# else:
if fred_api_key is None:
    print("FRED_API_KEY not found in environment variables.")

fred = Fred(api_key=fred_api_key)
```

```python
gdp = fred.get_series("GDP").shift(1).resample("QE").last()
gnp = fred.get_series("GNP").shift(1).resample("QE").last()
baa = fred.get_series("BAA").resample("QE").last()
aaa = fred.get_series("AAA").resample("QE").last()
cpi = fred.get_series("CPIAUCSL").resample("QE").last()
gs10 = fred.get_series("GS10").resample("QE").last()
tb3ms = fred.get_series("TB3MS").resample("QE").last()
```

```python
fred_df = pd.concat([gdp, gnp, baa, aaa, cpi, gs10, tb3ms], axis=1)
fred_df.columns = ["fred_gdp","fred_gnp","fred_baa","fred_aaa","fred_cpi","fred_gs10","fred_tb3ms"]
fred_df.index = fred_df.index.strftime('%Y-%m-%d')
```

```python
fred_df.to_csv("../data/fred_api_datasets_snapshot_" + today_str + ".csv")
fred_df.to_pickle("../data/fred_api_datasets_snapshot_" + today_str + ".pickle")
```
